```python
# Reading data set in csv format to pandas data frame.

dataframe = pd.read_csv("2022_02_08-02_30_31_AM.csv")
dataframe.head(5)
```

```python
product_with_prices = dataframe[dataframe["price_string"].notnull()]
product_without_prices = dataframe[dataframe["price_string"].isnull()]
```

Count of products with prices by product type

```python
product_with_prices.groupby("product_type")["uuid"].count()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_with_prices.groupby("product_type")["uuid"].count().index, product_with_prices.groupby("product_type")["uuid"].count().values).set(title="Count of products with prices by product type")
```

Count of products without prices by product type

```python
product_without_prices.groupby("product_type")["uuid"].count()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_without_prices.groupby("product_type")["uuid"].count().index, product_without_prices.groupby("product_type")["uuid"].count().values).set(title="Count of products without prices by product type")
```

Count of products with prices by level 1

```python
product_with_prices.groupby("level_1")["uuid"].count()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_with_prices.groupby("level_1")["uuid"].count().index, product_with_prices.groupby("level_1")["uuid"].count().values).set(title="Count of products with prices by level 1")
```

Count of products without prices by level 1

```python
product_without_prices.groupby("level_1")["uuid"].count()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_without_prices.groupby("level_1")["uuid"].count().index, product_without_prices.groupby("level_1")["uuid"].count().values).set(title="Count of products without prices by level 1")
```

Count of products with prices by category

```python
product_with_prices.groupby("category")["uuid"].count()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_with_prices.groupby("category")["uuid"].count().index, product_with_prices.groupby("category")["uuid"].count().values).set(title="Count of products with prices by categories")
```

Count of products without prices by category

```python
product_without_prices.groupby("category")["uuid"].count()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_without_prices.groupby("category")["uuid"].count().index, product_without_prices.groupby("category")["uuid"].count().values).set(title="Count of products without prices by categories")
```

Converting Price to decimal values

```python
product_with_prices["price_string"] = product_with_prices["price_string"].str.lstrip("$")
product_with_prices["price_string"] = product_with_prices["price_string"].astype('float')
product_with_prices = product_with_prices.rename(columns = {"price_string":"price"})
product_with_prices = product_with_prices.drop("price_string_unf", axis=1)
product_with_prices["currency"] = "$"
product_with_prices
```

Categories by average price of product

```python
product_with_prices.groupby("category")["price"].mean()
```

```python
sns.set(rc = {'figure.figsize':(30,8)})
sns.barplot(product_with_prices.groupby("category")["price"].mean().index, product_with_prices.groupby("category")["price"].mean().values).set(title="Mean of product prices by categories")
```
